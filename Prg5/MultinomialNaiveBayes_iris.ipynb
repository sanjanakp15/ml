The dataset tells us the features of different classes of iris. Here class is our target variables. We can see that all features influence the target.

Preprocessing is required as we need to convert the class of iris into numeric type using label encoding

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df.iloc[:,4]=le.fit_transform(df.iloc[:,4])
```

```python
df.head()
```

```python
x=df.drop('Class',axis=1)
y=df[['Class']]
```

New point

```python
test=[5,2.5,3,1.8]
```

train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=5)
```

```python
from sklearn.naive_bayes import MultinomialNB
multinb= MultinomialNB()
model= multinb.fit(x_train,y_train)
```

```python
y.value_counts()
```

We can see that the dataset is perfectly balanced

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
print(model.predict([test]))
```

Final Inference : Multinomial Naive Bayes algorithm is a probabilistic learning method. The algorithm is based on the Bayes theorem and predicts the target. Multinomial NB cares about counts for multiple features that do occur.
