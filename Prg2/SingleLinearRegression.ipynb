Plot the raw data and give label on both sides

```python
plt.scatter(data['SAT Score'], data.GPA, c='blue')
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.figure(figsize = (15,8))
```

The Dataset shows two columns which are the SAT Score and its GPA respectively.
We need to predict GPA for given SAT score.
Since the values are correct and are not NaN, no preprocessing is required.

Split into train and test data

```python
from sklearn.model_selection import train_test_split
x=data[['SAT Score']]
y=data[['GPA']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

Apply Linear Regression to find best fit

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

Print equation ,slope and intercept

```python
print("Slope(m): ",reg.coef_[0][0])
print("Intercept(c): ",reg.intercept_[0])
print("The Linear model is y = {:.2}*x + {:.3}".format(reg.coef_[0][0],reg.intercept_[0]))
```

Plot this best fit line on diff features


```python
predictions=reg.predict(x_test)
plt.figure(figsize=(16,8))
plt.scatter(x_test,y_test,c='red')
plt.plot(x_test,predictions,c='blue',linewidth=2)
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.show()
```

Find atleast 2 types of errors

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
rmse=np.sqrt(mean_squared_error(y_test,predictions))
mae=mean_absolute_error(y_test,predictions)
print("Root Mean Squared Error: ",rmse)
print("Mean Absolute Error: ",mae)
```

Reason for these errors : Differences between the actual and predicted values causes these errors

Final Inference : Linear Regression is a ML algorithm based on supervised learning. Regression models a target prediction value based on independent variable. Here we have SAT score as independent variable and GPA as target or dependent variable. We can calculate rmse and mae to check the accuracy of our models since these errors occur due to change in predicted value and actual value.
